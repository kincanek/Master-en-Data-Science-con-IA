## **FUNCIÓN DEFINIDA POR EL USUARIO**

Una **función definida por el usuario** es creada de manera personalizada para realizar tareas específicas y reutilizables en todo el programa. Este tipo de función se crea utilizando la palabra clave ` def` seguida del nombre de la función y los parámetros, si los hay.

Una función definida por el usuario tiene la siguiente estructura:

```
def nombre_funcion(parametro1, parametro2, ...):
    # Código que realiza una tarea específica

    # Devolver un valor (opcional)
    return valor
```

### **EJEMPLO 3:**

Función para calcular el factorial de un número:

````python
# difinir función
def factorial_(n):
    if n < 0:
        return None
    result = 1
    for i in range(1,n + 1):
        result *= i

    return result

print(factorial_(40))

# De manera recursiva
# Guarda cada llamada en la pila de ejecución
# hasta que llega al caso base, luego develve el resultado
def factorial(n):
    if n < 0:
        return None
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(40))
````

**3. Define una función que imprima todos los contactos de la agenda:**

   * Crea una función que realice un ciclo for que recorra y muestre cada contacto de la agenda.
   * Dentro del ciclo, muestra el número de contacto por cada iteración.

````python
# Crear función mostrar agenda
def print_agenda(agenda: dict, ordenar: bool = False) -> None:
    """
    Imprime los contactos de una agenda de forma formateada.

    Args:
        agenda (dict): Diccionario donde la clave es el nombre
            del contacto y el valor es otro diccionario con sus datos.
        ordenar (bool, optional): Si es True, ordena la agenda por nombre.
            Por defecto es False.
    """
    if not agenda:
        print("La agenda se encuentra vacía.")
        return

    items = sorted(agenda.items()) if ordenar else agenda.items()

    for index, (nombre, info) in enumerate(items, start=1):
        format_ = '=' * 15
        print(f"{format_} Contacto {index} {format_}")
        print(f"Nombre        : {nombre}")
        for clave, valor in info.items():
            print(f"{clave.capitalize():14}: {valor}")
        print() 
````

**7. Define una función para eliminar un contacto de la agenda:**

   * Crea una función que pida al usuario que ingrese el nombre del contacto a eliminar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces elimina el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".

````python
# Crear función eliminar contacto
def eliminar_contacto(agenda: dict,nombre: str) -> bool:
    """
    Elimina un contacto de la agenda.

    Args:
        agenda (dict): Diccionario de contactos.
        nombre (str): Nombre del contacto a eliminar.

    Returns:
        bool: True si el contacto fue eliminado correctamente,
              False si no existe o hubo error.
    """
    if agenda is None:
        print("No se ha proporcionado una agenda válida.")
        return False

    if not nombre.strip():
        print("Debe proporcionar un nombre válido.")
        return False

    if nombre not in agenda:
        print(f"El contacto '{nombre}' no existe.")
        return False

    del agenda[nombre]
    print(f"Contacto '{nombre}' eliminado correctamente.")
    return True
````

**8. Junta todas las funciones del menú CRUD dentro de una función:**

   * Crea una función crear un menú a través de un ciclo while.
   * Solicita al usuario una opción mostrada en el menú, si el usuario no coloca la opción correcta, entonces muestra el mensaje "Opción no encontrada".
   * Dada la opción escojida, ingresa a una función definida con anterioridad que corresponda a su operación a realizar.
   * Define en la última opción un `break` para salir del menú.

````python
while True:
    opcion_principal = menu_principal()

    if opcion_principal == 1:
        print("\n" + "=" * 70)
        print("AGENDA DE CONTACTOS")
        print("=" * 70)
        print_agenda(agenda)

    elif opcion_principal == 2:
        # Crear contacto
        print("\n" + "=" * 70)
        print("CREAR CONTACTO")
        print("=" * 70)
        print("Ingrese la información del nuevo contacto:\n")

        nombre = input("Nombre            : ").strip()
        telefono = input("Teléfono          : ").strip()
        correo = input("Correo electrónico: ").strip()

        crear_contacto(agenda, nombre, {"telefono": telefono, "correo": correo})

    elif opcion_principal == 3:
        # Consultar contacto
        print("\n" + "=" * 70)
        print("BUSCAR CONTACTO")
        print("=" * 70)

        nombre = input("Nombre del contacto a buscar: ").strip()
        buscar_contacto(agenda, nombre)

    elif opcion_principal == 4:
        # Actualizar contacto
        print("\n" + "=" * 70)
        print("ACTUALIZAR CONTACTO")
        print("=" * 70)

        nombre = input("Nombre del contacto a actualizar: ").strip()

        if not buscar_contacto(agenda, nombre):
            print("\n" + "-" * 70)
            print("No se puede actualizar: contacto no encontrado.")
            print("-" * 70)
            continue

        while True:
            opcion = menu_contacto()

            if opcion == 4:
                print("\n" + "-" * 70)
                print("Volviendo al menú principal...")
                print("-" * 70)
                break

            if opcion == 1:
                print("\n" + "-" * 70)
                print("ACTUALIZAR NOMBRE")
                print("-" * 70)

                nuevo_nombre = input("Nuevo nombre: ").strip()
                if actualizar_nombre(agenda, nombre, nuevo_nombre):
                    nombre = nuevo_nombre

            elif opcion == 2:
                print("\n" + "-" * 70)
                print("ACTUALIZAR TELÉFONO")
                print("-" * 70)

                nuevo_telefono = input("Nuevo teléfono: ").strip()
                actualizar_telefono(agenda, nombre, nuevo_telefono)

            elif opcion == 3:
                print("\n" + "-" * 70)
                print("ACTUALIZAR CORREO")
                print("-" * 70)

                nuevo_correo = input("Nuevo correo: ").strip()
                actualizar_correo(agenda, nombre, nuevo_correo)

    elif opcion_principal == 5:
        # Eliminar contacto
        print("\n" + "=" * 70)
        print("ELIMINAR CONTACTO")
        print("=" * 70)

        nombre = input("Nombre del contacto a eliminar: ").strip()
        eliminado = eliminar_contacto(agenda, nombre)

        if eliminado:
            print("\n" + "-" * 70)
            print("Agenda actualizada:")
            print("-" * 70)
            print_agenda(agenda)

    elif opcion_principal == 6:
        print("\n" + "=" * 70)
        print("Saliendo del programa... ¡Hasta luego!")
        print("=" * 70)
        break
````
