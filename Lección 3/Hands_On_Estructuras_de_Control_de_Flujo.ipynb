## **SELECCIÓN**

Una **estructura de selección** es una **declaración condicional**, el cual permite que diferentes secuencias de código se ejecuten dependiendo de si una **condición booleana** especificada se evalúa como **verdadera o falsa.**

La sintaxis básica de una condicional es la siguiente:

```
if condicion1:
    # código a ejecutar si la condicion1 es verdadera
elif condicion2: (opcional)
    # código a ejecutar si la condicion1 es falsa pero la condicion2 es verdadera
else:
    # código a ejecutar si tanto la condicion1 como la condicion2 son falsas
```

### **EJEMPLO 1:**

Determinar calificación de un examen:

````python
# Definir variable
calificacion = int(input('Inserta una calificación'))
````

````python
# Definir condicional
if calificacion == 100:
    print("Calificación perfecta")
elif calificacion >= 60 and calificacion < 100:
    print("Calificación aprobatoria")
elif calificacion >= 0 and calificacion < 60:
    print("Calificación aprobatoria")
else:
    print("Calificación no permitida")
````

## **ITERACIÓN**

En Python, una **estructura de iteración** es un **ciclo** que generalmente se refiere a dos tipos de **estructuras de control de flujo**: los bucles **for** y **while**.


*   **Bucle For**: El bucle for se usa para **recorrer sobre una secuencia** (que puede ser una lista, una tupla, un diccionario, un conjunto o una cadena) **o cualquier otro objeto iterable**. Cada elemento de la secuencia se procesa en el cuerpo del bucle, uno tras otro.

*   **Bucle While**: El bucle while se usa para **ejecutar repetidamente un bloque de código** siempre que una condición especificada sea verdadera.

### **CICLO FOR**

También conocido como **bucle for**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de código un número determinado de veces.

El ciclo for tiene la siguiente estructura:

```
for variable in secuencia:
    # cuerpo del ciclo

```

#### **EJEMPLO 1:**

Sumar los números del 1 a 10:

````python
# realiza el ciclo for
suma = 0
for i in range(10):
    suma += i

print(f'La suma es {suma}')
````

#### **EJEMPLO 2:**

Insertar numeros secuenciales dentro de una lista:

````python
# definir variables
secuencia = []
number_ = int(input('Inserta un número'))

for i in range(number_):
    secuencia.append(i + 1)

print(secuencia)
````

### **CICLO WHILE**

También conocido como **bucle while**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de código mientras se cumpla una condición específica.

El ciclo for tiene la siguiente estructura:

```
while condicion:
    # código a ejecutar mientras se cumpla la condición
```

#### **EJEMPLO 1:**

Validar una cadena:

````python
# realiza el ciclo while
bandera = True

while bandera:
    usuario = input("Ingrese su usuario: ").lower()
    if usuario == "root":
        print("Usuario correcto. Bienvenido.")
        bandera = False
    else:
        print("Usuario incorrecto. Inténtelo de nuevo.")
````

#### **EJEMPLO 2:**

Recorrer las opciones de un menú:

````python
# Definir variables
menu = """
****************MENU****************
1) Mi primer hola mundo
2) Factorial de 5
3) Salir
************************************
"""
print(menu)
````

````python
# Realizar ciclo while
while True:
    print(menu)
    opcion = input("Seleccione una opción: ")
    if opcion == "1":
        print("Hola Mundo")
    elif opcion == "2":
        factorial = 1
        for i in range(1, 6):
            factorial *= i
        print(f"El factorial de 5 es: {factorial}")
    elif opcion == "3":
        print("Saliendo del programa. ¡Hasta luego!")
        break
    else:
        print("Opción no válida. Por favor, intente de nuevo.")
````

#### **EJEMPLO 3:**

Calcular la suma de los valores de un diccionario:

````python
# define tus variables
diccionario = {"a": 1, "e": 2, "i": 3, "o": 4, "u": 5  }
vocal = input("Ingrese una vocal: ").lower()
claves =  list(diccionario.keys())
limite = claves.index(vocal)

lista = []
i = 0
suma = 0
while i <= limite:
    clave = claves[i]
    valor = diccionario[clave]
    lista.append(clave)
    suma += valor
    i += 1

print(f"La suma de estos valores es: {suma}")
print(f"Los valores son: {lista}")
````

**5. Actualiza un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre del contacto a actualizar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces imprime el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".
   * Crea un menú a través de un ciclo while que muestre las siguientes opciones:
   
   ```
   Menú actualizar contacto:
   
   1. Actualizar nombre
   2. Actualizar teléfono
   3. Actualziar correo
   4. Salir del menú
   ```
   * Solicita al usuario una opción mostrada en el menú, si el usuario no coloca la opción correcta, entonces muestra el mensaje "Contacto no encontrado".
   * Dada la opción escojida para modificar, ingresa ya sea el nuevo nombre, el nuevo teléfono o el nuevo correo del contacto encontrado.
   * Por cada registro nuevo ingresado, imprime de nueva forma el contacto para visualizar los cambios.
   * Define en la última opción un `break` para salir del menú.

````python
# Actualizar contacto
nombre_actualizar = input("Ingresa el nombre del contacto a actualizar: ")

contacto_encontrado = None
for nombre, datos in agenda.items():
    if nombre == nombre_actualizar:
        contacto_encontrado = nombre
        print("\nContacto encontrado:")
        print(f"Nombre: {nombre}")
        print(f"Teléfono: {datos['telefono']}")
        print(f"Correo: {datos['correo']}\n")
        break

if contacto_encontrado is None:
    print("Contacto no encontrado\n")
else:
    while True:
        print("Menú actualizar contacto:\n")
        print("1. Actualizar nombre")
        print("2. Actualizar teléfono")
        print("3. Actualziar correo")
        print("4. Salir del menú")

        opcion = input("\nElige una opción: ")

        if opcion == "1":
            nuevo_nombre = input("Ingresa el nuevo nombre: ")
            datos = agenda[contacto_encontrado]
            del agenda[contacto_encontrado]
            agenda[nuevo_nombre] = datos
            contacto_encontrado = nuevo_nombre

            print("\nContacto actualizado:")
            print(f"Nombre: {contacto_encontrado}")
            print(f"Teléfono: {agenda[contacto_encontrado]['telefono']}")
            print(f"Correo: {agenda[contacto_encontrado]['correo']}\n")

        elif opcion == "2":
            nuevo_tel = input("Ingresa el nuevo teléfono: ")
            agenda[contacto_encontrado]["telefono"] = nuevo_tel

            print("\nContacto actualizado:")
            print(f"Nombre: {contacto_encontrado}")
            print(f"Teléfono: {agenda[contacto_encontrado]['telefono']}")
            print(f"Correo: {agenda[contacto_encontrado]['correo']}\n")

        elif opcion == "3":
            nuevo_correo = input("Ingresa el nuevo correo: ")
            agenda[contacto_encontrado]["correo"] = nuevo_correo

            print("\nContacto actualizado:")
            print(f"Nombre: {contacto_encontrado}")
            print(f"Teléfono: {agenda[contacto_encontrado]['telefono']}")
            print(f"Correo: {agenda[contacto_encontrado]['correo']}\n")

        elif opcion == "4":
            break

        else:
            print("Contacto no encontrado\n")
````

**6. Elimina un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre del contacto a eliminar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces elimina el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".

````python
# Eliminar contacto
nombre_eliminar = input("Ingresa el nombre del contacto a eliminar: ")

eliminado = False
for nombre in list(agenda.keys()):
    if nombre == nombre_eliminar:
        del agenda[nombre]
        print("Contacto eliminado\n")
        eliminado = True
        break

if not eliminado:
    print("Contacto no encontrado\n")
````
