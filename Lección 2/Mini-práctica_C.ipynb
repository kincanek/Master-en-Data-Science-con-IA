Mini-práctica C (tipos de datos): pide 3 números al usuario, guárdalos en lista y calcula promedio; muestra el tipo de cada valor antes y después de convertir.

```python
dato_1 = input("Introduce el dato 1: ")
print(f'El dato introducido es {dato_1} y es de tipo {type(dato_1)}')
dato_2 = input("Introduce el dato 2: ")
print(f'El dato introducido es {dato_2} y es de tipo {type(dato_2)}')
dato_3 = input("Introduce el dato 3: ")
print(f'El dato introducido es {dato_3} y es de tipo {type(dato_3)}')

datos = [float(dato_1), float(dato_2), float(dato_3)]
print(f'La suma de los datos es: {sum(datos)}')
print(f'El promedio de los datos es: {sum(datos)/len(datos)}')
```
