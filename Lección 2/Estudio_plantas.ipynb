# Muestras de medidas de tallos de plantas

Base de datos de zonas de muestreo con diferentes plantas.

```python
datos ={"P1": {"Altura": {
        "datos": [ 12,13,10],
        "unidad": "cm" },
        "Grosor": {"datos": [11,10,11],
        "unidad": "mm"}},
    "P2": {
      "Altura": {
        "datos": [7.6,9.9,12,11.3,8.4,7.5],
        "unidad": "cm"},
        "Grosor": {
        "datos": [8.25,10,9.14,9.15,8.2,9.7],
        "unidad": "mm"}},
    "P3": {
      "Altura": {
        "datos": [10,8,10,10.5],
        "unidad": "cm"},
        "Grosor": {
        "datos": [11.22,14.53,14.14,9.59],
        "unidad": "mm"}}}
        
```

```python
print("El promedio de altura de la planta P1 es: ", sum(datos["P1"]["Altura"]["datos"])/len(datos["P1"]["Altura"]["datos"]), datos["P1"]["Altura"]["unidad"])
print("El promedio de grosor de la planta P1 es: ", sum(datos["P1"]["Grosor"]["datos"])/len(datos["P1"]["Grosor"]["datos"]), datos["P1"]["Grosor"]["unidad"])
```

```python
print('El promedio de altura de todas las plantas es: ', (sum(datos["P1"]["Altura"]["datos"])+sum(datos["P2"]["Altura"]["datos"])+sum(datos["P3"]["Altura"]["datos"]))/(len(datos["P1"]["Altura"]["datos"])+len(datos["P2"]["Altura"]["datos"])+len(datos["P3"]["Altura"]["datos"])), "cm")
```
